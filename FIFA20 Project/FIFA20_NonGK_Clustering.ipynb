## DataFrame Potential Position

```python
def cleaning(x):
    if x[i]==0:
        return 0
    elif type(x[i])!= str:
        return x[i]
    else :
        return x[i].split(sep="+")[0]
```

```python
df_stats = df.iloc[:,79:-1]
for i in df_stats.columns:
    df_stats[i]= df_stats.apply(lambda x:cleaning(x),axis=1)
df_stats
```

```python
df_stats['average potential skill']=df_stats.mean(axis=1) 
df_stats
```

# KMEANS

```python
from sklearn.cluster import KMeans
inertia = []

for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df_final.values)
    inertia.append(kmeans.inertia_)



plt.figure(figsize=(20, 10))
# plt.plot(inertia)

sns.lineplot(x=range(2, 11), y=inertia, color='#000087', linewidth = 4)
sns.scatterplot(x=range(2, 11), y=inertia, s=300, color='#800000',  linestyle='--')
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(df_final.values)
```

# VISUALIZATION

```python
from sklearn.decomposition import PCA 
pca = PCA(n_components=2)

pca.fit(df_final)
pcs = pca.transform(df_final)

data_pca = pd.DataFrame(data = pcs, columns = ['PC_1', 'PC_2'])
data_pca['clusters'] = kmeans.labels_
data_pca.head()
```

```python
fig, ax = plt.subplots(figsize=(15,10))

sns.scatterplot(
    x="PC_1", y="PC_2",
    hue="clusters",
    #edgecolor='green',
    linestyle='--',
    data=data_pca,
    palette=['blue','red','yellow','black'],
    s=160,
    ax=ax
)
```

```python
from sklearn.decomposition import PCA 
pca = PCA(n_components=3)

pca.fit(df_final)
pcs = pca.transform(df_final)

data_pca = pd.DataFrame(data = pcs, columns = ['PC_1', 'PC_2','PC_3'])
data_pca['clusters'] = kmeans.labels_
data_pca.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt



fig = plt.figure(figsize=(20,15))
ax = fig.add_subplot(111, projection='3d')

x =data_pca['PC_1']
y =data_pca['PC_2']
z =data_pca['PC_3']



ax.scatter(x, y, z, c=data_pca['clusters'], marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')


plt.show()
```

```python
from sklearn.metrics import silhouette_score
for n_clusters in range (2, 11):
    clusterer = KMeans(n_clusters=n_clusters)
    preds = clusterer.fit_predict(df_final)
    centers = clusterer.cluster_centers_

    score = silhouette_score(df_final, preds)
    print("For n_clusters = {}, silhouette score is {})".format(n_clusters, score))
```
