# Import Library and Data Extraction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('DataFrameTeamPositionGK_AllClub.csv')
```

```python
df
```

## Dataframe Based On Physique

```python
physic_df = df[['age','height_cm','weight_kg','overall','potential','preferred_foot','weak_foot','body_type','player_traits']]
```

```python
physic_df.head()
```

```python
le = LabelEncoder()
physic_df['preferred_foot'] = le.fit_transform(physic_df['preferred_foot'])
physic_df['body_type']      = le.fit_transform(physic_df['body_type'])
physic_df['player_traits']  = df.player_traits.str.count(',')
physic_df.head()
```

```python
physic_df_mean = pd.DataFrame(physic_df.mean(axis=1),columns=['average physical mean'])
physic_df_mean
```

## DataFrame Based on General Stats

```python
df_gk = df.iloc[:,38:44]
df_gk = df_gk.fillna(0)
df_gk['average goalkeeping skill']=df_gk.mean(axis=1)
df_gk
```

```python
gk =df_gk[['average goalkeeping skill']]
gk
```

## DataFrame Based On Attacking

```python
df_attack = df.iloc[:,45:50]
df_attack = df_attack.fillna(0)
df_attack['average attacking skill']=df_attack.mean(axis=1)
attack = df_attack[['average attacking skill']]
attack
```

## DataFrame Based on Skill

```python
df_skill = df.iloc[:,50:55]
df_skill = df_skill.fillna(0)
df_skill['average controling skill']=df_skill.mean(axis=1)
skill = df_skill[['average controling skill']]
skill
```

## DataFrame Based on Power

```python
df_pow = df.iloc[:,60:65]
df_pow = df_pow.fillna(0)
df_pow['average power skill'] = df_pow.mean(axis=1)
power = df_pow[['average power skill']]
power
```

## DataFrame Based on Mental

```python
df_mental = df.iloc[:,65:71]
df_mental = df_mental.fillna(0)
df_mental['average mentality skill']=df_mental.mean(axis=1)
mental = df_mental[['average mentality skill']]
mental
```

# DataFrame Based on Defense

```python
df_def =df.iloc[:,71:74]
df_def=df_def.fillna(0)
df_def['average defense skill']= df_def.mean(axis=1)
defense_mean = df_def[['average defense skill']]
defense_mean
```

# Combine Each DataFrame

```python
Final = pd.concat([physic_df_mean,defense_mean,mental,power,move,skill,attack,gk],axis=1)
Final
```

```python
ss = StandardScaler()
Final2 = ss.fit_transform(Final)
Final_df = pd.DataFrame(Final2,columns=Final.columns)
Final_df
```

# KMEANS

```python
from sklearn.cluster import KMeans
inertia = []

for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(Final_df.values)
    inertia.append(kmeans.inertia_)



plt.figure(figsize=(20, 10))
# plt.plot(inertia)

sns.lineplot(x=range(2, 11), y=inertia, color='#000087', linewidth = 4)
sns.scatterplot(x=range(2, 11), y=inertia, s=300, color='#800000',  linestyle='--')
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(Final_df.values)
```

# VISUALIZATION

```python
from sklearn.decomposition import PCA 
pca = PCA(n_components=2)

pca.fit(Final_df)
pcs = pca.transform(Final_df)

data_pca = pd.DataFrame(data = pcs, columns = ['PC_1', 'PC_2'])
data_pca['clusters'] = kmeans.labels_
data_pca.head()
```

```python
fig, ax = plt.subplots(figsize=(15,10))

sns.scatterplot(
    x="PC_1", y="PC_2",
    hue="clusters",
    #edgecolor='green',
    linestyle='--',
    data=data_pca,
    palette=['blue','red','yellow','black'],
    s=160,
    ax=ax
)
```

```python
from sklearn.decomposition import PCA 
pca = PCA(n_components=3)

pca.fit(Final_df)
pcs = pca.transform(Final_df)

data_pca = pd.DataFrame(data = pcs, columns = ['PC_1', 'PC_2','PC_3'])
data_pca['clusters'] = kmeans.labels_
data_pca.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt



fig = plt.figure(figsize=(20,15))
ax = fig.add_subplot(111, projection='3d')

x =data_pca['PC_1']
y =data_pca['PC_2']
z =data_pca['PC_3']



ax.scatter(x, y, z, c=data_pca['clusters'], marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

## Silhouette Score

```python
from sklearn.metrics import silhouette_score
for n_clusters in range (2, 11):
    clusterer = KMeans(n_clusters=n_clusters)
    preds = clusterer.fit_predict(Final_df)
    centers = clusterer.cluster_centers_

    score = silhouette_score(Final_df, preds)
    print("For n_clusters = {}, silhouette score is {})".format(n_clusters, score))
```

# Use the best 2 skill of GK Players from Final dataframe

```python
GoalkeepingAndMovement = FinalBest4[['average movement skill','average goalkeeping skill']]
GoalkeepingAndMovement = GoalkeepingAndMovement.fillna(0)
GoalkeepingAndMovement['GoalkeepingAndMovement'] = GoalkeepingAndMovement.mean(axis=1)
GKandMovementMean = GoalkeepingAndMovement[['GoalkeepingAndMovement']]
GKandMovementMean
```

```python
PhysicalAndPower = FinalBest4[['average physical mean','average movement skill']]
PhysicalAndPower = PhysicalAndPower.fillna(0)
PhysicalAndPower['PhysicalAndPower'] = PhysicalAndPower.mean(axis=1)
PhysicAndPower = PhysicalAndPower[['PhysicalAndPower']]
PhysicAndPower
```

```python
best2skill = pd.concat([GKandMovementMean, PhysicAndPower],axis=1)
best2skill
```

```python
from sklearn.cluster import KMeans
inertia = []

for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(best2skill.values)
    inertia.append(kmeans.inertia_)



plt.figure(figsize=(20, 10))
# plt.plot(inertia)

sns.lineplot(x=range(2, 11), y=inertia, color='#000087', linewidth = 4)
sns.scatterplot(x=range(2, 11), y=inertia, s=300, color='#800000',  linestyle='--')
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(best2skill)
```

```python
best2skill['cluster'] = kmeans.labels_
best2skill
```
