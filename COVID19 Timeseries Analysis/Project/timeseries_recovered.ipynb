# AR Model

## Split the Dataset into Train set and Test Set

In this dataset, I will separate the dataset based on Monthly Data

1. Train Dataset = Data of **January 2020-July 2020**
2. Test Dataset = Data of **August 2020-September 2020**

### Find 01/08/2020 Index Data

```python
multiple_diff_recoveredCopy = multiple_diff_recovered.copy()
multiple_diff_recoveredCopy.reset_index(inplace=True)
multiple_diff_recoveredCopy.columns=['Datetime', 'WorldRecovered_Total']
multiple_diff_recoveredCopy
```

```python
AugustIndex = multiple_diff_recoveredCopy.loc[multiple_diff_recoveredCopy['Datetime']=='8/1/20']
AugustIndex
```

Based on above, the index of **1st August 2020 is 190**

## Create AR Model

```python
train = multiple_diff_recovered[:190]
test = multiple_diff_recovered[190:]
```

```python
train.head()
```

```python
test.head()
```

```python
test.info()
```

```python
from statsmodels.tsa.ar_model import AR
import numpy as np

model_ar = AR(np.asarray(train))
```

```python
model_ar_fit = model_ar.fit()
```

```python
predict = model_ar_fit.predict(start=190, end=241)
```

```python
import matplotlib.pyplot as plt

plt.plot(test)
plt.plot(predict)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(test, predict)
```

## Moving Average Model

```python
multiple_diff_recovered
```

```python
multiple_diff_recovered['MA_10'] = multiple_diff_recovered.WorldRecovered_Total.rolling(10, min_periods=1).mean()
multiple_diff_recovered['MA_20'] = multiple_diff_recovered.WorldRecovered_Total.rolling(20, min_periods=1).mean()
```

```python
multiple_diff_recovered
```

```python
colors = ['yellow', 'blue', 'red']

multiple_diff_recovered.plot(color=colors, figsize=(15,12))

plt.legend(labels=['WorldRecovered_Total','MA_10','MA_20'])
```
