# 1. Indonesia Confirmed COVID19 Cases

```python
import pandas as pd
```

```python
dfConfirmed = pd.read_csv('time_series_covid19_confirmed_global.csv')
dfConfirmed
```

```python
dfConfirmed['TotalConfirmed'] = dfConfirmed.sum(axis=1)
dfConfirmed
```

```python
dfConfirmed['Country/Region'].value_counts()
```

# Combine and Sum new dataframe with the same country but has different province

## Get list of columns that we want to sum

```python
listColumnsConfirmed = list(dfConfirmed.columns)
listColumnsConfirmed = listColumnsConfirmed[4:]
listColumnsConfirmed
```

## Create new DataFrame with groupby of Country/Region and Sum all of it

```python
dfConfirmed = dfConfirmed.groupby(['Country/Region'])[listColumnsConfirmed].sum().reset_index()
dfConfirmed
```

```python
dfConfirmed['Country/Region'].value_counts()
```

```python
dfIndonesiaConfirmed = dfConfirmed[dfConfirmed['Country/Region']=='Indonesia']
dfIndonesiaConfirmed
```

```python
dfIndonesiaConfirmed = pd.DataFrame(dfIndonesiaConfirmed.sum())
dfIndonesiaConfirmed.columns = ['IndonesiaConfirmed_Total']
dfIndonesiaConfirmed = dfIndonesiaConfirmed.iloc[1:-1, : ]
dfIndonesiaConfirmed
```

```python
dfIndonesiaConfirmed_diff = dfIndonesiaConfirmed.diff(periods=1)
dfIndonesiaConfirmed_diff = dfIndonesiaConfirmed_diff[1:]
dfIndonesiaConfirmed_diff
```

```python
dfIndonesiaConfirmed_diff.plot()
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(dfIndonesiaConfirmed_diff)
```

```python
dfIndonesiaConfirmed_multiplediff = dfIndonesiaConfirmed_diff.diff(periods=1)
dfIndonesiaConfirmed_multiplediff = dfIndonesiaConfirmed_multiplediff.iloc[1:]
dfIndonesiaConfirmed_multiplediff
```

```python
dfIndonesiaConfirmed_multiplediff.plot()
```

```python
plot_acf(dfIndonesiaConfirmed_multiplediff)
```

# Create AR Model for Indonesia Confirmed Data

```python
train_IndonesiaConfirmed = dfIndonesiaConfirmed_multiplediff[:190]
test_IndonesiaConfirmed = dfIndonesiaConfirmed_multiplediff[190:]
```

```python
train_IndonesiaConfirmed.head()
```

```python
test_IndonesiaConfirmed.head()
```

```python
from statsmodels.tsa.ar_model import AR
import numpy as np

modelAR_IndonesiaConfirmed = AR(np.asarray(train_IndonesiaConfirmed))
```

```python
modelAR_IndonesiaConfirmed_fit = modelAR_IndonesiaConfirmed.fit()
```

```python
predictIndonesiaConfirmed = modelAR_IndonesiaConfirmed_fit.predict(start=190, end=241)
```

```python
import matplotlib.pyplot as plt

plt.plot(test_IndonesiaConfirmed)
plt.plot(predictIndonesiaConfirmed)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(test_IndonesiaConfirmed, predictIndonesiaConfirmed)
```
