# AR Model

## Split the Dataset into Train Set and Test Set

In this dataset, I will separate the dataset based on Monthly Data

1. Train Dataset = Data of **January 2020-July 2020**
2. Test Dataset = Data of **August 2020-September 2020**

### Find 01/08/2020 Index Data

```python
multiple_diff_DeathCopy = multiple_diff_Death.copy()
multiple_diff_DeathCopy.reset_index(inplace=True)
multiple_diff_DeathCopy.columns=['Datetime', 'WorldDeath_Total']
multiple_diff_DeathCopy
```

```python
AugustIndex = multiple_diff_DeathCopy.loc[multiple_diff_DeathCopy['Datetime']=='8/1/20']
AugustIndex
```

Based on above, the index of **1st August 2020 is 190**

# Create AR Model

```python
train = multiple_diff_Death[:190]
test = multiple_diff_Death[190:]
```

```python
train.head()
```

```python
test.head()
```

```python
test.info()
```

```python
from statsmodels.tsa.ar_model import AR
import numpy as np

model_ar = AR(np.asarray(train))
```

```python
model_ar_fit = model_ar.fit()
```

```python
predict = model_ar_fit.predict(start=190, end=241)
```

```python
import matplotlib.pyplot as plt

plt.plot(test)
plt.plot(predict)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(test, predict)
```
