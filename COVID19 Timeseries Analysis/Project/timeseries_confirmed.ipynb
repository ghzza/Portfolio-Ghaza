# AR Model

## Split the Dataset into Train Set and Test Set

In this dataset, I will separate the dataset based on Monthly Data.

1. Train Dataset = Data of **January 2020 - July 2020**
2. Test Dataset = Data of **August 2020 - September 2020**

### Find 01/08/2020 Index Data

```python
multiple_diff_ConfirmedCopy = multiple_diff_Confirmed.copy()
multiple_diff_ConfirmedCopy.reset_index(inplace=True)
multiple_diff_ConfirmedCopy.columns= ['Datetime','WorldConfirmed_Total']
multiple_diff_ConfirmedCopy
```

```python
AugustIndex = multiple_diff_ConfirmedCopy.loc[multiple_diff_ConfirmedCopy['Datetime'] == '8/1/20']
AugustIndex
```

Based on above, the index of **1st August 2020 is 190**

## Create AR Model

```python
train = multiple_diff_Confirmed[:190]
test = multiple_diff_Confirmed[190:]
```

```python
train.head()
```

```python
test.head()
```

```python
test.info()
```

```python
from statsmodels.tsa.ar_model import AR
import numpy as np

model_ar = AR(np.asarray(train))
```

```python
model_ar_fit = model_ar.fit()
```

```python
predict = model_ar_fit.predict(start=190, end=241)
```

```python
import matplotlib.pyplot as plt

plt.plot(test)
plt.plot(predict)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(test, predict)
```

## Moving Average Model

```python
multiple_diff_Confirmed
```

```python
multiple_diff_Confirmed['MA_10']=multiple_diff_Confirmed.WorldConfirmed_Total.rolling(10,min_periods=1).mean()
multiple_diff_Confirmed['MA_20']=multiple_diff_Confirmed.WorldConfirmed_Total.rolling(20,min_periods=1).mean()
```

```python
multiple_diff_Confirmed
```

```python
colors = ['yellow', 'blue','red']

multiple_diff_Confirmed.plot(color=colors, figsize=(15,12))

plt.legend(labels=['WorldConfirmed_Total','MA_10','MA_20'])
```
